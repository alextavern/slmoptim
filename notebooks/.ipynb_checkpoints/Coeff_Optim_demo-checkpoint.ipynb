```python
import sys, time
import numpy as np
import pickle
from tqdm import tqdm, trange
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
from aotools.functions import phaseFromZernikes, zernike_noll

sys.path.append(str('/home/optomeca/Projects'))

from slmOptim.loader import patterns as pt
from slmOptim.loader.patterns import GaussPatternGenerator, ZernikePatternGenerator, superpose, LaguerrePatternGenerator, PlaneWaveGenerator
from slmOptim.initializer.init_experiment import InitExperiment
from slmOptim.optimizer.iter import CoefficientsOptimization
from slmOptim.utils import cost_func

```

### Define parameters

```python

# Define observation window on camera
roi_size = 600
off = (-10, 20)
off = (-20, 50)
# Define camera macropixel, this is defined by the measured speckel grain
bins = 1

# Define camera acquisition parameters
exposure_time = 10
gain = 1
timeout= 100

# path to save raw data
path = '/home/optomeca/Documents/measurements'
```

### Initialize experiment 
#### camera and slm

```python

init = InitExperiment(roi_size, off, bins, exposure_time, gain, timeout)
camera = init.init_cam()
# slm = init.init_slm()

def set_mirror(slm=slm, resolution=(800, 600)):
    resX, resY = resolution
    patSLM = pt.PatternsBacic(resX, resY)
    mirror = patSLM.mirror()
    slm.sendArray(mirror)
    time.sleep(.2)
```

check the mode to optimize

```python
set_mirror()
# slm.sendArray(mask)
frame = camera.get_pending_frame_or_null()
frame = np.copy(frame.image_buffer)
plt.imshow(frame)
plt.colorbar()

print(frame.shape)

# init.close_cam()
```

```python
init.close_cam()
```

#### Define cost function
Two cost functions are tested. One maximises the energy on a given circle on the frame and one maximized the correlation coefficient between the frame and a (gaussian) target spot. 

```python
# define cost function and pass it to the object
gauss = GaussPatternGenerator(num_of_segments=604 * 604, num_of_patterns=1, w0=0.5e-3) 
amp = gauss[0]
cost_fun = lambda x: cost_func.corr_coef(x, amp)
# cost_fun = lambda x: cost_func.snr(x, 4, (0, 0))

plt.imshow(amp)
```

```python
# define how many polynomials to optimize
num = 50

pattern_loader = ZernikePatternGenerator(600 * 600, num, phase=False)
# pattern_loader = PlaneWaveGenerator(600, 112,  krange=(1, 3, 1), phistep=10)
# pattern_loader = GaussPatternGenerator(600 * 600, num, phase=False, w0=0.5e-3, LG=True)
# pattern_loader = LaguerrePatternGenerator(num_of_segments=600 * 600, num_of_patterns=num, waist=500)

supermask = lambda x: superpose(pattern_loader, x)
```

#### Optimize

```python
# reset slm
set_mirror()

# instantiate zernike class
zern = CoefficientsOptimization(slm, camera, (600, 800), num_of_coeffs=num, save_path=path)

# pass cost function to object
zern.register_cost_callback(cost_fun)
# pass patterns
zern.register_pattern_callback(supermask)

# run algorithm
optim = zern.run((-2, 2, 0.25))

# save raw date if you want
zern.save_raw()

# load optimized mask to SLM
key = list(optim['masks'].keys())[-1]
mask = optim['masks'][key]
slm.sendArray(mask)
```

### Do some plotting and save

```python
coeffs = optim['coeffs']
cost = optim['cost']
frames = optim['frames']
masks = optim['masks']

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(8, 4))
msk = axs[0].imshow(mask)
axs[0].set_title("Optimized mask")
axs[0].set_xlabel("SLM x-pixels #")
axs[0].set_ylabel("SLM y-pixels #")
divider = make_axes_locatable(axs[0])
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(msk, cax=cax)

axs[1].plot(coeffs)
axs[1].set_title("Optimized coefficients")
axs[1].set_xlabel("Number of iterations #")
axs[1].set_ylabel("Coefficients")
axs[1].set_box_aspect(1)


axs[2].plot(cost)
axs[2].set_title("Cost function vs iterations")
axs[2].set_xlabel("Number of iterations #")
axs[2].set_ylabel("Cost function")
axs[2].set_box_aspect(1)

fig.tight_layout()

figpath = zern.filepath + 'optim'
plt.savefig(figpath, dpi=200, transparent=True)

def filter_dict(dict, num, N):
    z0 = 1
    z1 = num
    keys =list(np.linspace(z0, z1, N).astype('uint8'))
    dict_new = {key:dict[key] for key in keys}
    return dict_new

def plots(dict, mn=(3, 5), figsize=(8, 4), log=False):
    nrows, ncols = mn
    fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize)

    for (idx, matrix), ax in zip(dict.items(), axs.ravel()):
        if log:
            im = ax.imshow(matrix, norm=LogNorm())
        else:     
            im = ax.imshow(matrix)
        ax.axis('off')
        ax.set_title(r"$Z_{%d}$" %(idx))
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        fig.colorbar(im, cax=cax)
    fig.tight_layout()

    return fig
    

modes = filter_dict(frames, num, 15)
fig1 = plots(modes, log=False, mn=(3, 5), figsize=(8, 4))
figpath = zern.filepath + 'optim_modes'
plt.savefig(figpath, dpi=200, transparent=True)


modes = filter_dict(frames, num, 15)
fig1 = plots(modes, log=True, mn=(3, 5), figsize=(8, 4))
figpath = zern.filepath + 'optim_modes_log'
plt.savefig(figpath, dpi=200, transparent=True)

masks = filter_dict(masks, num, 15)
fig2 = plots(masks, mn=(3, 5), figsize=(8, 4))
figpath = zern.filepath + 'optim_masks'
plt.savefig(figpath, dpi=200, transparent=True)

    
```
